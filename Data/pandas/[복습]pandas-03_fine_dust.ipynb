```python
### 라이브러리 사용하는 것 import 
import pandas as pd
import numpy as np
```

```python
# [3-1] '2016년 ~ 2019년 미세먼지' 파일을 가져와 df2016 ~ df2019라는 이름을 부여합니다.
df2016=pd.read_excel('./data/data_day_2016.xlsx',engine='openpyxl')
df2017=pd.read_excel('./data/data_day_2017.xlsx',engine='openpyxl')
df2018=pd.read_excel('./data/data_day_2018.xlsx',engine='openpyxl')
df2019=pd.read_excel('./data/data_day_2019.xlsx',engine='openpyxl')
```

```python
# [3-2] 미세먼지 통합 데이터를 concat을 이용해서 행으로 합한다.
# 모든 컬럼이 같으므로 axis=0으로, 기존 인덱스번호를 버리기 위해 ignore_index=True준다
fine_dust = pd.concat([df2016,df2017,df2018,df2019],axis= 0,ignore_index=True)
fine_dust.info()
```

```python
# [3-2-1] : ignore_index 대신 index를 만들어서 사용가능
fine_dust.index = pd.RangeIndex(len(fine_dust))
fine_dust.info()
```

```python
# [3-3] 파일이 존나 많을 경우에는? glob 모듈을 사용 
import glob
file_list= glob.glob('./data/data_day_*.xlsx') # 이렇게 생긴 애들을 전부 불러다 준다
fine_dust =[]
for dust in sorted(file_list):
    temp = pd.read_excel(dust)
    fine_dust.append(temp)
fine_dust = pd.concat(fine_dust,axis=0)
fine_dust.shape
```

```python
# [3-4] df의 row의 개수 및 각 column의 정보 및 메모리 사용량을 확인
print(df2016.shape,df2017.shape,df2018.shape,df2019.shape)
```

```python
# [3-18] '측정일시2' 컬럼이 필요 없으므로 제거합니다. axis를 정해주면 얘가 잘찾음
fine_dust.drop('측정일시2',axis=1).head(2)
```

```python
fine_dust = pd.read_csv('data/fine_dust.csv')
fine_dust.head()
```

```python
# [3-19] 결측치를 확인함
fine_dust.isna().sum(axis=0) # 0은 생략가능함
```

```python
# [3-20] '측정일시'를 index로 설정하고, 시간을 기준으로 오름차순 정렬
# 그래프에서 y축으로 사용하려고 합니다.
fine_dust2 = fine_dust.set_index('측정일시').sort_index()
fine_dust2.head()
```

```python
# [3-21] Feature Engineering : '측정소명', '년', '월', '일','오존농도(ppm)', '미세먼지(㎍/㎥)', '초미세먼지(㎍/㎥)'컬럼만
# 추출하여 df_dust 라는 이름을 정의합니다.
df_dust = fine_dust2[['측정소명','year','month','day','오존농도(ppm)','미세먼지(㎍/㎥)','초미세먼지(㎍/㎥)']]
df_dust.head(2)
```

```python
# [3-22] df_dust의 index를 다시 컬럼으로 사용되도록 합니다. 
df_dust= df_dust.reset_index()
df_dust.head(2)
```

### 결측치 제거
1. DataFrame.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
   - 결측치 제거에 사용되는 메서드
   - how='any' : 결측치가 하나라도 포함된 행 삭제
   - how='all' : 모든 데이터가 결측치인 행 삭제
   - axis=1 : 컬럼에 대해 동작
   - thresh=숫자 : 숫자 이상의 데이터를 가진 행은 삭제 안함
   - subset=[컬럼이름1, ...] : subset으로 지정된 컬럼만 사용하여 삭제 대상 검색

```python
# [3-23] df_dust 에서 ['오존농도(ppm)','미세먼지(㎍/㎥)', '초미세먼지(㎍/㎥)']에서
# 모든 데이터가 결측치인 행을 제거하여 결과를 temp1으로 저장합니다
temp = df_dust.dropna(how='all',subset=['오존농도(ppm)','미세먼지(㎍/㎥)', '초미세먼지(㎍/㎥)'])

# df_dust와 temp1의 행의 수 (len)을 확인하여 제거된 행의 개수를 확인합니다.
print(len(df_dust)-len(temp))
```

```python
# [3-24] df_dust 에서 ['오존농도(ppm)','미세먼지(㎍/㎥)', '초미세먼지(㎍/㎥)']에서
# 하나라도 결측치가 있는 행을 제거하여 결과를 temp2로 저장합니다.
temp2 = df_dust.dropna(how='any',subset=['오존농도(ppm)','미세먼지(㎍/㎥)', '초미세먼지(㎍/㎥)'])

# df_dust와 temp2의 행의 수 (len)을 확인하여 제거된 행의 개수를 확인합니다.
print(len(df_dust)-len(temp2))
```

```python
# [3-25] df_dust 에서 ['오존농도(ppm)','미세먼지(㎍/㎥)', '초미세먼지(㎍/㎥)']에서
# 2개 이상의 데이터를 가진 행은 제거하지 않은 결과를 temp3로 저장합니다.
# (= 3개의 정보 중 1개의 데이터만 가진 행을 제거함)
temp3 = df_dust.dropna(thresh=2,subset=['오존농도(ppm)','미세먼지(㎍/㎥)', '초미세먼지(㎍/㎥)'])

# df_dust와 temp3의 행의 수 (len)을 확인하여 제거된 행의 개수를 확인합니다.
print(len(df_dust)-len(temp3))
```

```python
# [3-26] df_dust에서 '년', '월'별 '미세먼지(㎍/㎥)' 데이터의 평균을 구해
# DataFrame으로 만들어 meandf 라는 이름을 지정합니다.ㅣ
meandf = df_dust.groupby(['year','month'])[['미세먼지(㎍/㎥)']].mean()
meandf
```

```python
# [3-27] meandf에서 2017년 6월까지의 데이터만 출력
# loc할때 튜플형태로 하면 순서대로 인덱싱 
meandf.loc[:(2017,6)]
```

```python
# [3-28] meandf에 '결측치제거후' 및 '차이'라는 컬럼을 추가합니다.
# '결측치제거후' 컬럼은 temp2에서 '년', '월'별 '미세먼지(㎍/㎥)' 데이터의 평균을 사용합니다.
# '차이' 컬럼은 '미세먼지(㎍/㎥)' - '결측치제거후' 를 사용합니다.
meandf['결측치제거후']=temp2.groupby(['year','month'])['미세먼지(㎍/㎥)'].mean()
meandf['차이']=meandf['미세먼지(㎍/㎥)'] - meandf['결측치제거후']
meandf
```

```python
# [3-29] meandf에서 2017년 6월까지의 데이터만 출력합니다.
# 여러 개의 값에 차이가 있는 것을 확인할 수 있음
meandf[:(2017,6)]
```

### 결측치 대체 (평균값)
> 결측치를 제거하는 것은 큰 위험이 도사리므로, 대체하는 것을 더 긍정적으로 검토해봐야 함
1. Series.mask(조건, 조건이 참일 때 사용할 값)
  - s.isna() : NA값에 대해 True, NA아닌 것은 False
2. Series.where(조건, 조건이 거짓일 때 사용할 값 또는 값 목록)
  - s.notna() : NA값에 대해 False, NA아닌 것은 True
