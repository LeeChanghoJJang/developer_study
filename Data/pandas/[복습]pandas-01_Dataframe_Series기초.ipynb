### loc, iloc 사용하기
1. df.iloc[행,열] : 행과 열에 순서(`번호`)를 기준으로 데이터들을 조회(슬라이싱 가능)
2. df.loc[행,열] : 행과 열의 이름(`문자`)를 기준으로 데이터 조회(슬라이싱 가능)
   - 중요한 것은 조건을 넣을 수도 있으며, &와 | 둘다 가능
3. isin : 특정 데이터값이 들어있는지 여부를 True, False로 반환

### 데이터 값 변경 방법
- Series.replace(변경대상, 변경내용)
- Series.replace([변경대상1, 변경대상2, ...], [변경내용1, 변경내용2, ...])
- Series.replace({변경대상1:변경내용1, 변경대상2:변경내용2, ...})
- replace는 기본적으로 값 전체를 변경대상으로 지정함
- regex=True 를 사용하면 변경대상을 일부 내용만 대상으로 지정 할 수 있음
- regex => regular expression(정규식, 정규표현식)

```python
# [1-12] temp의 필드의 데이터 값을 먼저 변경합니다.
# replace할 때, 해당 컬럼에 변경된 내용을 저장시켜줘야 함.
# regex가 True로 되어야 값의 일부만이라도 수정할 수 있음.
temp['거래금액(만원)'] = temp['거래금액(만원)'].replace(',','',regex=True)
# 마찬가지로 데이터 타입이 변경된 경우에도 해당 컬럼에 변경된 내용으로 저장
temp['거래금액(만원)'] = temp['거래금액(만원)'].astype('int')
temp['거래금액(만원)'].dtype
```

### datetime, category 타입변경
1. pd.to_datetime(Series, format='형식') 또는 DateFrame.astype('datetime64[ns]')
    - %Y: 4글자 년도, %y : 2글자 년도, %m : 2글자 월, %d : 2글자 일
    - 다양한 형식 문자들 : https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior
    - format의 지정 필수 X
2. pd.Categorical(Series, categories=['범주1', '범주2', ...], ordered=None)
    - ordered=True 사용시 순서 있는 범주형, 정렬시 정해진 순서가 사용됨
    - categories, ordered의 지정이 필수는 아님


```python
# [1-17] datetime과 category 타입 변경을 위한 데이터프레임 정의
temp = pd.DataFrame({'날짜_일반': ['2021/01/01', '2021/01/02', '2021/01/03', '2021/01/04', '2021/01/05'],
                     '날짜_시간': ['2021-01-01 1:12:10', '2021-01-02 1:13:45', '2021-01-03 2:50:10', '2021-01-04 3:12:30', '2021-01-05 5:40:20'],
                     '날짜_특수': ['21-01-01', '21-01-02', '21-01-03', '21-01-04', '21-01-05'],
                     '범주': ['금', '토', '일', '월', '화']})
temp.info()
```

```python
# [1-18] datetime 타입 변경을 위해 astype 메소드 활용('YY/MM/DD'의 형태는 날짜로 변경가능)
# 결과물은 (YY-MM-DD) 형태로 출력됨(np.datetime64 대신 datetime64[ns] 사용할 것)
t1=temp['날짜_일반'].astype('datetime64[ns]')
t1.dtype
```

```python
# [1-19] 시간까지 나와있는 날짜의 dtype을 변경
# 결과물은 시간까지 다 나옴 
t2 = temp['날짜_시간'].astype('datetime64[ns]')
t2.dtype
```

```python
# [1-20] 날짜_특수의 dtype을 변경
t3= temp['날짜_특수'].astype('datetime64[ns]')
t3.dtype
```

```python
# [1-21] pd.to_datetime(Series, format='형식')을 사용하여 '날짜_특수'의 dtype을 변경해 보자
t4 =pd.to_datetime(temp['날짜_특수'],format='%y-%m-%d')
t4
```

```python
# [1-22] Series.astype('category')를 사용하여 '범주'의 dtype 변경
s5=temp['범주'].astype('category')
s5
```

```python
# [1-23] 위에서 만들어진 '범주'를 정렬해 보자 (Series.sort_values() 사용)
s5.sort_values()
```

```python
# [1-24] pd.Categorical(Series, categories=카테고리목록, ordered=True)를 사용해 요일 순 범주를 만들어 보자
# 요일목록 => ['월', '화', '수', '목', '금', '토', '일']
s6= pd.Categorical(temp['범주'],categories=['월', '화', '수', '목', '금', '토', '일'],ordered=True)
s6.sort_values() # 다시 sort_values 하면 월화수목금토일 순으로 출력됨
```

### 통계값 알아보기
- DataFrame.describe() : 숫자형 데이터타입을 갖는 컬럼에 대해 count, mean, std, min, max, Q1, Q2, Q3 등의 통계값 구하기
- Series.count() : 개수
- Series.sum() : 합계
- Series.mean() : 평균
- Series.std() : 표준편차
- Series.var() : 분산
- Series.median() : 중앙값
- Series.mode() : 최빈값
- Series.quantile([0.5, 0.75]) : 분위수

```python
## [1-27] 거래금액의 합계, 개수, 평균, 표준편차, 분산, 중앙값, 최빈값 도출
df1000 = df['거래금액(만원)'].replace(',','',regex=True).astype('int')
df1000.mean()
df1000.std()
df1000.var()
df1000.median()
df1000.mode()
df1000.describe()
```

### 그룹별 통계치 구하기

- DataFrame.groupby(그룹명)[[컬럼1, 컬럼2, ...]].적용할통계함수() : DataFrame

```python
# [1-28] 시군구별 건축년도의 평균을 구합니다.
df123 = df.groupby('시군구')['건축년도'].mean()
df123
```
